## Step 2: Clean your data.

##### Do some rudimentary cleaning. Rename any columns that are poorly named, shorten any strings that may be too long, fill in any missing values. 

```python
## Renaming the columns 
characters = {'th': "", 'x':"", 'st':"", 'nd': "", 'rd':"", '.':"_", '*': '', '!':''}
new_columns = { } 

for search, j in characters.items():
#     print search, j
#     print music.columns
    for index, col in enumerate(music.columns):
#         print col, type(col), col.replace(search, j)
        if col in new_columns.keys():
            new_columns[col] = new_columns[col].replace(search,j)
        else:
            new_columns[col] = col.replace(search, j)
new_columns
music.rename(columns=new_columns, inplace=True) 
music.columns.values
```

```python
## Tabing exploring columns
for column in music:
      print (music[column].value_counts())
# year looks good
# arti_inverted looks good
# track looks good
# time looks good
```
